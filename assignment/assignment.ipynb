```python
import pandas as pd
# 1
df = pd.read_csv('./data/college_completion.csv')
```

```python
# 2
df.describe()
# There are 3798 observations included in the dataset, with 50 variables used.
```

```python
df.head()
```

```python
# 3
pd.crosstab(df['control'], df['level'])

```

There appear to be very few private non-profit two-year colleges included, while that category is much better represented for four-year institutions. Private non-profit colleges make up the largest single category for four-year institutions, while the majority of two-year institutions included are public.

```python
# 4
import seaborn as sns
sns.histplot(df['grad_100_value'])
```

```python
sns.kdeplot(df['grad_100_value'])
```

```python
df['grad_100_value'].describe()
```

```python
# 5

import matplotlib.pyplot as plt
sns.kdeplot(df, x='grad_100_value', hue='control',  common_norm=False)
plt.title('Kernel Density Plot of Graduation Rate by Control')
plt.xlabel('Graduation Rate')
plt.ylabel('Density')
plt.show()
```

```python
sns.kdeplot(data=df, x='grad_100_value', hue='level',  common_norm=False)
plt.title('Kernel Density Plot of Graduation Rate by Level')
plt.xlabel('Graduation Rate')
plt.ylabel('Density')
plt.show()
```

```python
df['grad_100_value'].groupby(df['control']).describe()
```

```python
df['grad_100_value'].groupby(df['level']).describe()
```

It appears that public and 2-year institutions have the best graduation rates in their respective variables. This likely means that community colleges have the best graduation rates, being 2-year public colleges. This likely stems from the fact that they are both affordable and are just 2 years in length, making the associates degree easier to finish.

```python
# 6
df['levelXcontrol'] = df['level']+', '+df['control']

sns.kdeplot(df, x = 'grad_100_value', hue = 'levelXcontrol')
```

As observed from the previous two graphs, 2-year public institutions have the highest graduation rates. 4-year private for-profit colleges also have high graduation rates. 

```python
# 7

sns.kdeplot(df, x = 'aid_value', hue = 'level')
plt.xlabel('Average Amount of Aid')
plt.title('Amount of Aid by Type of College(Length)')
```

```python
sns.kdeplot(df, x = 'aid_value', hue = 'control')
plt.title("Amount of Aid by Type of College(Length)")
plt.xlabel("Average Amount of Aid")
plt.show()
```

```python
df['aid_value'].groupby(df['control']).describe()
```

```python
df['grad_100_value'].groupby(df['level']).describe()
```

```python
# 8
sns.scatterplot(df, x = 'grad_100_value', y = 'aid_value')
```

There appears to be an upward trend between increased aid and graduation rate, at least for a section of colleges.

```python
sns.scatterplot(df, x = 'grad_100_value', y = 'aid_value', hue = 'control')
```

```python
sns.scatterplot(df, x = 'grad_100_value', y = 'aid_value', hue = 'level')
```

Colleges that are private non-profit and 4 years tend to see the strongest correlation between financial aid and graduation rate, while 2 year public colleges see little increase of aid in coordination with graduation rate.
